# Running Assignment in Colab

The following is a script which lets you interact with the .py files in the assignment entirely from within colab. 

The workflow is straightforward: 

1. Mount GoogleDrive
2. Navigate to the repo 
3. Pull any recent changes 
4. Run main

The following variables should be set by you to your path (all updated in the 
first code block): 

* REPOPATH: Where the repo is stored on GoogleDrive
* REPONAME: The name of the repo
* GITEMAIL: Email for github
* GITNAME: Name for github
